```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example data
data = sns.load_dataset('iris')['sepal_length']

# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=False, bins=30)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# KDE Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data, shade=True)
plt.title('KDE Plot')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Value')
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(data)
plt.title('Violin Plot')
plt.xlabel('Value')
plt.show()
```

```python
# Strip Plot
plt.figure(figsize=(10, 6))
sns.stripplot(x=data)
plt.title('Strip Plot')
plt.xlabel('Value')
plt.show()
```
