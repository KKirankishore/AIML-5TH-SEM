```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame
df = sns.load_dataset('titanic')

# Pie Chart for the 'class' variable
class_counts = df['class'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Passenger Class')
plt.show()
```

```python
# Bar Plot for the 'embarked' variable
plt.figure(figsize=(10, 6))
sns.countplot(x='embarked', data=df)
plt.title('Bar Plot of Embarked Ports')
plt.xlabel('Port of Embarkation')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd

# Stacked Bar Plot for 'class' and 'sex'
class_sex_counts = df.groupby(['class', 'sex']).size().unstack()
class_sex_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Stacked Bar Plot of Passenger Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Sex')
plt.show()
```

```python
# Donut Chart for the 'class' variable
class_counts = df['class'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))
plt.title('Donut Chart of Passenger Class')
plt.show()
```

```python
# Area Chart for the 'class' variable (example using continuous data)
df['class_num'] = df['class'].map({'First': 1, 'Second': 2, 'Third': 3})
class_freq = df['class_num'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.fill_between(class_freq.index, class_freq, step="mid", alpha=0.4)
plt.plot(class_freq.index, class_freq, marker='o')
plt.title('Area Chart of Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```
