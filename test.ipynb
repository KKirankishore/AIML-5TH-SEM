```python
import pandas as pd

data = {
    'ename': ['rashmi', 'amar', 'ashwini', 'lohit', 'mohan', 'pramod'],
    'Type': ['regular', 'adhoc', 'regular', 'adhoc', 'contract', 'regular'],
    'department': ['cs', 'cs', 'ec', 'me', 'civil', 'ec'],
    'experience': [10, 20, 5, 14, 7, 8],
    'salary': [50000, 15000, 30000, 15000, 10000, 40000]
}

df = pd.DataFrame(data)
print(df)
```

```python
pivot_avg_salary = df.pivot_table(values='salary', index='Type', aggfunc='mean')
print(pivot_avg_salary)
```

```python
import matplotlib.pyplot as plt
plt.bar(df['Type'],df['salary'])
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.pie(df['salary'])
plt.show()
```

```python
import pandas as pd
data = {
    'name': ['john', 'vihan', 'krishna', 'arthi'],
    'maths': [75, 75, 96, 87],
    'science': [95, 50, 65, 50],
    'social': [50, 95, 56, 60],
    'english': [99, 79, 96, 81],
    'hindi': [85, 35, 76, 89]
}
df = pd.DataFrame(data)
print(df)
```

```python
pivot_avg_scores = df.pivot_table(values=['maths', 'science', 'social', 'english', 'hindi'], index='name', aggfunc='mean')
print(pivot_avg_scores)
```

```python
df['total'] = df[['maths', 'science', 'social', 'english', 'hindi']].sum(axis=1)
print(df)
```

```python
sorted_df = df[df['maths'] > 80].sort_values(by='maths', ascending=False)
print(sorted_df)
```

```python
import matplotlib.pyplot as plt

df.plot(x='name', y=['maths', 'science', 'social', 'english', 'hindi'], kind='bar')
plt.xlabel('Scores')
plt.ylabel('name')
plt.title('Student Scores by Subject')
plt.legend(loc='upper right')
plt.show()
```
