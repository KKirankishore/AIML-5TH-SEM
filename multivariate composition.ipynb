```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create sample data
dates = pd.date_range(start='2023-01-01', periods=10, freq='D')
data = pd.DataFrame({
    'Date': dates,
    'Category A': np.random.rand(10),
    'Category B': np.random.rand(10),
    'Category C': np.random.rand(10)
})

# Set 'Date' as the index
data.set_index('Date', inplace=True)

# Plot
plt.figure(figsize=(10, 6))
plt.stackplot(data.index, data['Category A'], data['Category B'], data['Category C'],
              labels=data.columns, alpha=0.6)

plt.title('Stacked Area Plot')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend(loc='upper left')
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Category 1', 'Category 2', 'Category 3']
variables = ['Variable A', 'Variable B', 'Variable C']
data = np.array([
    [5, 10, 15],
    [6, 12, 18],
    [7, 14, 21]
])

# Create a stacked bar plot
fig, ax = plt.subplots()

# Set up the positions and width for the bars
bar_width = 0.5
index = np.arange(len(categories))

# Plot each variable as a separate bar segment
bottoms = np.zeros(len(categories))
for i, variable in enumerate(variables):
    ax.bar(index, data[:, i], bar_width, bottom=bottoms, label=variable)
    bottoms += data[:, i]

# Add labels and title
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Stacked Bar Plot for Multivariate Composition')
ax.set_xticks(index)
ax.set_xticklabels(categories)
ax.legend(title='Variables')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey

# Create a Sankey diagram object
sankey = Sankey(unit=None)

# Add flows: (flow, label, orientation, width)
sankey.add(flows=[10, -10, 5, -5], labels=['Input 1', 'Output 1', 'Input 2', 'Output 2'])

# Create the plot
fig, ax = plt.subplots()
sankey.finish()
ax.title.set_text('Sankey Plot with Matplotlib')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the outer pie chart
outer_labels = ['Category A', 'Category B', 'Category C']
outer_sizes = [30, 40, 30]
outer_colors = ['#ff9999','#66b3ff','#99ff99']

# Data for the inner pie chart
inner_labels = ['Subcategory A1', 'Subcategory A2', 'Subcategory B1', 'Subcategory B2', 'Subcategory C1', 'Subcategory C2']
inner_sizes = [15, 15, 20, 20, 10, 20]
inner_colors = ['#ff6666','#ff9999','#66c2a5','#99d3a2','#c2c2f0','#b3b3b3']

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(aspect="equal"))

# Outer pie chart
outer_pie, _ = ax.pie(outer_sizes, labels=outer_labels, colors=outer_colors, radius=1, wedgeprops=dict(width=0.3, edgecolor='w'))

# Inner pie chart
inner_pie, _ = ax.pie(inner_sizes, labels=inner_labels, colors=inner_colors, radius=0.7, wedgeprops=dict(width=0.3, edgecolor='w'))

plt.title('Nested Pie Chart Example')
plt.show()
```
