```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

# Creating a DataFrame 
df = pd.DataFrame({
    "date": pd.date_range(start="1990-09-07", periods=25, freq="YE"),
    "temp": np.random.randint(18, 30, size=25)
})

print(df)

# Plot the DataFrame
plt.figure(figsize=(10, 5))
ax = df.plot(kind='bar', x='date', y='temp', color='r', linewidth=1, figsize=(10, 5))
plt.xticks(rotation=25)
ax.set_ylabel('Temperature')
ax.set_xlabel('Date')
plt.title('Temperature over Time')
plt.grid(True)
plt.show()
```
