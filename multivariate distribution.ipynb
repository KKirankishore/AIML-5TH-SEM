```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate some data
np.random.seed(0)
x = np.random.randn(10000)
y = np.random.randn(10000)

# Create a hexbin plot
plt.figure(figsize=(10, 6))
plt.hexbin(x, y, gridsize=50, cmap='Blues')

# Add color bar
cb = plt.colorbar(label='counts')

# Add labels
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.title('Hexbin Plot')

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate a sample DataFrame with multivariate data
np.random.seed(42)
data = pd.DataFrame({
    'A': np.random.normal(0, 1, 100),
    'B': np.random.normal(5, 2, 100),
    'C': np.random.normal(-3, 1, 100),
    'D': np.random.normal(7, 2, 100)
})

# Create a pairplot to visualize the scatter plots and distribution of variables
sns.pairplot(data, diag_kind='kde')  # 'kde' for Kernel Density Estimation on the diagonal
plt.suptitle('Pairplot of Multivariate Distribution', y=1.02)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Example dataset
data = sns.load_dataset('iris')

# Create a pair plot
sns.pairplot(data, hue='species')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example dataset for Anscombe's quartet
df = sns.load_dataset("iris")

# Create a joint plot
sns.jointplot(data=df, x="sepal_length", y="sepal_width", kind="scatter")

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate some random multivariate data
np.random.seed(0)
mean = [0, 0, 0]
cov = [[1, 0.8, 0.7], [0.8, 1, 0.6], [0.7, 0.6, 1]]  # Covariance matrix
data = np.random.multivariate_normal(mean, cov, 100)

# Split the data into x, y, z coordinates
x, y, z = data.T

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z, c='b', marker='o')

# Add labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Show plot
plt.show()
```
