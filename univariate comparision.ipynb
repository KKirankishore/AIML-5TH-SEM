```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame with categories
df = sns.load_dataset('titanic')  # Example dataset

# Box Plot for Age by Class
plt.figure(figsize=(12, 6))
sns.boxplot(x='pclass', y='age', data=df)
plt.title('Box Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
# Violin Plot for Age by Class
plt.figure(figsize=(12, 6))
sns.violinplot(x='pclass', y='age', data=df)
plt.title('Violin Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
# Strip Plot for Age by Class
plt.figure(figsize=(12, 6))
sns.stripplot(x='pclass', y='age', data=df, jitter=True)
plt.title('Strip Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
# Swarm Plot for Age by Class
plt.figure(figsize=(12, 6))
sns.swarmplot(x='pclass', y='age', data=df)
plt.title('Swarm Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
# KDE Facet Grid for Age by Class
g = sns.FacetGrid(df, col='pclass', col_wrap=3, height=4)
g.map(sns.kdeplot, 'age', fill=True)
g.set_axis_labels('Age', 'Density')
g.set_titles('Passenger Class {col_name}')
plt.show()
```
